1)Create a scatter plot to visualize the relationship between two variables, by generating a synthetic
dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.rand(100) * 10
y = 2.5 * x + np.random.randn(100) * 5
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='teal', alpha=0.7, edgecolor='black')
plt.title('Synthetic Scatter Plot: X vs Y')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.grid(True)
plt.tight_layout()
plt.show()
```

2)Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='steelblue', edgecolor='black', alpha=0.8)
plt.title('Distribution of Randomly Generated Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

3)Create a dataset representing categories and their corresponding values. Compare different categories
based on numerical values.

```python
import matplotlib.pyplot as plt

categories = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
values = [120, 90, 150, 110, 130]

plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.title('Comparison of Product Sales')
plt.xlabel('Product Category')
plt.ylabel('Sales (Units)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

4)Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.



```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
categories = ['Category A', 'Category B', 'Category C']
data = [
    np.random.normal(50, 10, 100),
    np.random.normal(60, 15, 100),
    np.random.normal(55, 20, 100)
]

plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=categories, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='black'),
            medianprops=dict(color='red'),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'))
plt.title('Distribution of Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

5)Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
size = 100
x1 = np.random.normal(0, 1, size)
x2 = x1 * 0.8 + np.random.normal(0, 0.2, size)
x3 = x1 * -0.6 + np.random.normal(0, 0.3, size)
x4 = np.random.normal(0, 1, size)
df = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, square=True, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()

```
